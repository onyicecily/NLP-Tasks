```python
import requests
from bs4 import BeautifulSoup
from gensim.summarization import summarize
```

```python
url = 'https://www.npr.org/2021/04/28/991683886/frightened-to-death-cheerleader-speech-case-gives-supreme-court-pause'
```

```python
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page)
```

```python
header = soup.find('h1').text
```

```python
date = soup.find('span', attrs={'class':'date'}).text
```

```python
p_tags = soup.find_all('p')
p_tags
```

```python
text = [tags.text.strip() for tags in p_tags]
text
```

```python
sentence_list = [sentence for sentence in text if not '\n' in sentence]
```

```python
sentence_list = [sentence for sentence in sentence_list if '.' in sentence]
```

```python
article = ''.join(sentence_list)
```

```python
summary = summarize(article)
```

```python
summary
```

# Tokenization

```python
import spacy
```

### Load the language model

```python
nlp = spacy.load('en_core_web_sm')
```

<img src = 'nlp.png'>

```python
doc = nlp(summary)
```

```python
spacy.explain?
```

```python
for token in doc:
    print(token.text, '--->', spacy.explain(token.text))
```

# POS tagging

```python
for token in doc:
    print(token.text, '--->', token.pos_,'--->', spacy.explain(token.pos_))
```

# Dependency parsing

```python
for token in doc:
    print(token.text, '--->', token.dep_, '--->', spacy.explain(token.dep_))
```

# Lemmatization

```python
for token in doc:
    print(token.text, '--->', token.lemma_)
```

# Sentence boundary detection

```python
sentences = list(doc.sents)
```

```python
for sentence in sentences:
    print(sentence)
```

# Named entity 

```python
for ent in doc.ents:
    print(ent.text, '---->', ent.label_, '---->', spacy.explain(ent.label_))
```

# Visualization

```python
from spacy import displacy
```

```python
displacy.render(doc, style='ent', jupyter=True)
```

```python
displacy.render(doc, style='dep', jupyter=True)
```

# Putting it all together

```python
lis = []
```

```python
for token in doc:
    dic = {}
    dic['Token'] = token.text
    dic['POS'] = token.pos_
    dic['Tags'] = token.tag_
    dic['Dep'] = token.dep_
    dic['Explanation'] = spacy.explain(token.tag_)
    lis.append(dic)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(lis)
```

```python
data
```

```python
l = []
for ent in doc.ents:
    d = {}
    d['entities'] = ent.text
    d['labels'] = ent.label_
    d['Explanation'] = spacy.explain(ent.label_)
    l.append(d)
```

```python
entities = pd.DataFrame(l)
```
